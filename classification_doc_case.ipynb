```python
import pandas as pd
import nltk
```

```python
data = pd.read_csv("classificatie_opleidingen_train.csv")
```

```python
data
```

```python
# first, we'll remove the stop words from dutch language
```

```python
# Import stopwords with nltk.
from nltk.corpus import stopwords
stop = stopwords.words('dutch')
```

```python
# eliminate the stop words in dutch for the columns, and create "new columns without these words (these columns' name end with _without_stopwords)"

data['omschr_opleidingshoofdcategorie_nl_without_stopwords'] = data['omschr_opleidingshoofdcategorie_nl'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
data['omschr_opleidingssubcategorie_nl_without_stopwords'] = data['omschr_opleidingssubcategorie_nl'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
data['benaming_without_stopwords'] = data['benaming'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))


```

```python
data
```

```python
# function that eliminates punctuation marks in the dataframe columns
import string 
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text
```

```python
data["omschr_opleidingshoofdcategorie_nl_without_stopwords"] = data['omschr_opleidingshoofdcategorie_nl_without_stopwords'].apply(remove_punctuations)
data["omschr_opleidingssubcategorie_nl_without_stopwords"] = data['omschr_opleidingssubcategorie_nl_without_stopwords'].apply(remove_punctuations)
data["benaming_without_stopwords"] = data['benaming_without_stopwords'].apply(remove_punctuations)
```
